# Task 2 â€” Predict Future Stock Prices (Short-Term)
Goal: Use historical Apple stock data to predict the next day's closing price using regression.

Dataset: Yahoo Finance (AAPL, last 2 years).

Model Used: Random Forest Regressor

Features: Open, High, Low, Volume

Target: Close price of the next day

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Using yfinance to download Apple (AAPL) stock data for the last 2 years
# df contains columns: Open, High, Low, Close, Volume, etc.

import yfinance as yf

stock = yf.Ticker("AAPL")

df = stock.history(period="2y")
```

```python
# Display the first few rows to understand the structure of the dataset
df.head()
```

```python
# Shows data types, non-null values, and memory usage
df.info()
```

```python
# Check number of rows and columns in the dataset
df.shape
```

```python
# Check if the dataset contains missing values in any column
df.isnull().sum()
```

```python
# Check if the dataset contains duplicated values in any column
df.duplicated().sum()
```

```python
# Scatterplots for visualizations
plt.figure(figsize=(12,5))
sns.scatterplot(data=df,x='Date',y='Close')
plt.title("AAPL Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(data=df,x='Date',y='Volume')
plt.title("AAPL Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
# X contains all input features except 'Close'
# y is the target we want to predict: next day's closing price
X = df.drop('Close',axis=1)
y = df['Close']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
model = RandomForestRegressor(random_state=101)
```

```python
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
# Plotting Actual vs Predicted closing prices to visually evaluate model performance
plt.figure(figsize=(6,6))
sns.scatterplot(x=y_test, y=preds,hue=df.loc[y_test.index,'Close'] ,palette="viridis", alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Predicted vs Actual (Hue = Close)")
plt.show()
```

```python
# Evaluating model performance using:
# MAE (Mean Absolute Error)
# MSE (Mean Squared Error)
# RÂ² Score (how well the model fits the data)

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("Test MAE:", mean_absolute_error(y_test, preds))
print("Test MSE:", mean_squared_error(y_test, preds))
print("Test R^2:", r2_score(y_test, preds))
```

# Final Insights:
-The model learned patterns from 2 years of Apple stock data.

-Scatter plot shows how closely predictions follow actual closing prices.

-MAE/MSE indicate the typical prediction error in dollars.

-Random Forest performed well for short-term movement prediction, but stock prices remain highly volatile and not perfectly predictable.
